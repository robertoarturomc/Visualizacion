## Alternativas a Graficos de Pastel

### Graficos de Barras Apilados (Stacked bar Plots)

Estos son bastante utiles cuando tengo mas de un grupo. Por ejemplo, si quiero ver las ventas tptales por producto a lo largo de distintos años.


```python
años = ["2018", "2019", "2020", "2021", "2022"]
pan_blanco = [53, 55, 45, 48, 49]
pan_integral = [38, 45, 35, 55, 58]
```

```python
plt.bar(años, pan_blanco)

#plt.bar(años,  pan_integral, bottom =pan_blanco)
plt.title("Ventas Anuales de Pan (en Millones)")

plt.show()
```

Por cierto, tambien es posible agregar titulos en los ejes:

```python
plt.bar(años, pan_blanco)
plt.bar(años,  pan_integral, bottom =pan_blanco)
plt.title("Ventas Anuales de Pan (en Millones)")

plt.xlabel("Año")
plt.ylabel("Venta Total")

plt.show()
```

Y un gran consejo es agregar una **leyenda**, para saber qué significan los distintos colores.

```python
plt.bar(años, pan_blanco)
plt.bar(años,  pan_integral, bottom =pan_blanco)
plt.title("Ventas Anuales de Pan (en Millones)")
plt.xlabel("Año")
plt.ylabel("Venta Total")

# Debe haber una etiqueta para cada uso de plt.bar
plt.legend(["Blanco", "Integral"])

plt.show()
```

Esta se pone por default en el "mejor" lugar posible (donde estorbe menos).

### ¿Y si quiero, en vez de los datos en valores absolutos, en valores relativos?

Bueno, solo tengo que cambiar mis datos para que esten en porcentajes.

```python
pan_blanco, pan_integral = np.array(pan_blanco), np.array(pan_integral)

pan_blanco_p = pan_blanco / (pan_blanco + pan_integral)
pan_integral_p = pan_integral / (pan_blanco + pan_integral)

print(pan_blanco_p)
print(pan_integral_p)
```

```python
plt.bar(años, pan_blanco_p)
plt.bar(años,  pan_integral_p, bottom =pan_blanco_p)
plt.title("Ventas Anuales de Pan (en Millones)")
plt.xlabel("Año")
plt.ylabel("Venta Total")

plt.legend(["Blanco", "Integral"])

plt.show()
```

## Grafico de Area

Este grafico funciona similar al de barras apiladas. No obstante, cuando existe un orden entre los datos que van en el eje X (por ejemplo, al usar años), puede ser mejor idea este grafico, para observar si existen tendencias.

```python
plt.stackplot(años, pan_blanco, pan_integral)

plt.title("Ventas Anuales de Pan (en Millones)")
plt.xlabel("Año")
plt.ylabel("Venta Total")

plt.legend(["Blanco", "Integral"])

plt.show()
```

Y como siempre, puedo jugar con los parámetros para tener visualizaciones con puntos de vista alternativos.

```python
plt.stackplot(años, pan_blanco, pan_integral, baseline="sym")

plt.title("Ventas Anuales de Pan (en Millones)")
plt.xlabel("Año")
plt.ylabel("Venta Total")

plt.legend(["Blanco", "Integral"])

plt.show()
```

## Gráficos de Waffle

Estos Graficos utilizan areas para visualizar las cantidades. Esto funciona mejor que los graficos de circulo porque el ser humano es mucho mejor para identificar diferencias entre largos y anchos, que ángulos.

```python
# Hay que descargar la libreria pywaffle
!pip install pywaffle
```

```python
# Posteriormente, hay que importar la funcion Waffle
from pywaffle import Waffle
```

```python
# Datos
valores = [65, 123, 48]

# Waffle chart
plt.figure(
    FigureClass = Waffle, # El tipo de figura
    rows = 15, # Cuantas hileras
    values = valores) # Qué valores se grafican

plt.title("Alumnos del Departamento")

plt.show() 
 
```

¡Recuerda que es buena idea agregar una leyenda! En estos casos, hay que cambiar los datos a un formato de diccionario.

```python
valores = {"CDD": 65, "Actuaria" : 123, "Fisica": 48}

plt.figure(
    FigureClass = Waffle,
    rows = 15,
    values = valores,
    legend={"loc":"upper center"}) # Puedo especificar donde quiero la leyenda.

plt.title("Alumnos del Departamento")
plt.show() 
 
```

E incluso, puedo ponerme creativo y cambiar los cuadrados por alguna forma.

```python
plt.figure(
    FigureClass = Waffle,
    rows = 15,
    values = valores,
    legend={"loc":"upper center"},
    icons = 'user') 

plt.title("Alumnos del Departamento")
plt.show() 

```
