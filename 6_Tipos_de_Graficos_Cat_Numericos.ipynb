```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Tipos de Graficos
## Variables Numéricas separadas por Variables Categóricas.

En algunas ocasiones, nos interesa ver la distribución de alguna variable numérica, pero separada según alguna variable categórica.

Por ejemplo, en tu última tarea hiciste algo así:

```python
iris = sns.load_dataset("iris")

sns.displot(iris, x="petal_length", kind="kde", hue = "species")
plt.title("Distribucion de la longitud del pétalo por especie")
plt.show()
```

De esta forma, podemos ver la distribucion de `petal_Length` de acuerdo a las tres especies distintas.

Es más, podríamos tener una planta del género _iris_ pero, al no ser biólogos, no podríamos identificarla correctamente. No obstante, sabemos medir, y podemos ver la longitud de los pétalos. 

Gracias a la gráfica anterior, podría clasificar la planta dentro de alguna de las tres especies con bastante exactitud (y no fue necesario crear un modelo de *Machine Learning*).

Hoy veremos algunas otras alternativas que me permiten visualizar esto mismo. 

(Por cierto, un histograma también me permite ver lo anterior, pero no es taaan efectivo, ya que se enciman las barras).

```python
sns.displot(iris, x="petal_length", kind="hist", hue = "species")
plt.title("Histograma de la longitud del pétalo por especie")
plt.show()
```

## Diagrama de Caja (Diagrama de Caja y Bigotes)

Permiten ver, de una forma simplificada, la distribución de una variable. 

La "Caja" representa los cuartiles (percentil 25 y 75), la línea de enmedio, la mediana (percentil 50) y los bigotes, los valores mínimo y máximo. Si hubieran valores extremos (1.5 * el Rango Intercuartil <Q3 - Q1>), estos se grafican por separado.

![Boxplot](https://i.ytimg.com/vi/BE8CVGJuftI/maxresdefault.jpg)

* Ayudan a detectar valores candidatos a ser atípicos (extremos, *outliers*)
* Nos dicen si la variable es simétrica: para ello, la mediana tiene que estar enmedio del rectángulo.
* Simplifican el análsis.

```python
plt.boxplot([iris.loc[iris.species=="setosa", "petal_length"],
             iris.loc[iris.species=="versicolor", "petal_length"],
             iris.loc[iris.species=="virginica", "petal_length"]],
              labels=["setosa", "versicolor", "virginica"])
plt.title("Boxplot de la longitud de pétalo según especie")
plt.show()
```

Este gráfico también se puede hacer con seaborn (y más rápido, de hecho).

```python
sns.catplot(data=iris, x="species", y="petal_length", kind="box")
plt.title("Boxplot de la longitud de pétalo según especie - seaborn")
plt.show()
```

## Strip Plots / Swarm Plots

Son variantes de los *scatterplots* (gráficos de puntos/dispersión), pero agrupando según alguna variable categórica. Si los puntos no son demasiados, se pueden generar estos gráficos. De lo contrario, una posible solución es hacer un muestreo.

Tienen la ventaja de ser más explícitos que los boxplots.

```python
sns.catplot(data=iris, x="species", y="petal_length", kind="strip")
plt.title("Strip plot de la longitud de pétalo según especie")
plt.show()
```

```python
sns.catplot(data=iris, x="species", y="petal_length", kind="swarm")
plt.title("Swarm plot de la longitud de pétalo según especie")
plt.show()
```

La diferencia entre ambos es que los Swarm plots generan *bins* (como en los histogramas) para agrupar los datos.

Por cierto, ¿cómo le hago para hacer énfasis en algún punto en particular?
Una forma es agregando texto en algún punto para identificar algo.

```python
sns.catplot(data=iris, x="species", y="petal_length", kind="swarm")
plt.title("Strip plot de la longitud de pétalo según especie")

# Función de matplotlib para agregar text
plt.text(x=1, y=3, s="Valor atipico")
plt.show()
```

## Gráfico de Violín

Son la combinación entre un KDEplot (gráfico de densidad) y un boxplot. Permiten ver la distribución de cada variable de una forma suavizada.

```python
sns.catplot(data=iris, x="species", y="petal_length", kind="violin")
plt.title("Gráfico de Violín de la longitud de pétalo según especie")
plt.show()
```
